### Funciones básicas

```python
import numpy as np

datos = [2, 4, 6, 8, 10]

promedio = np.mean(datos)

print(promedio)
```

```python
import numpy as np

datos = [2, 4, 6, 8, 10]

varianza = np.var(datos)

print(varianza)
```

```python
import numpy as np

datos = [2, 4, 6, 8, 10]

desv_est = np.std(datos)

print(desv_est)
```
