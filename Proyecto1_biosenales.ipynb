### *Consultar historia, uso y proyectos futuros relacionados con Interfaz Cerebro Computador*

Las interfaces cerebro-computador (BCI, por sus siglas en inglés) son sistemas que permiten establecer una comunicación directa entre el cerebro y dispositivos externos sin la participación del sistema neuromuscular. Estas tecnologías se basan en la adquisición y procesamiento de señales cerebrales, comúnmente mediante electroencefalografía (EEG), con el fin de traducir la actividad neuronal en comandos que pueden ser utilizados para controlar sistemas electrónicos.

Los fundamentos de las BCI se remontan a 1924, cuando Hans Berger realizó el primer registro de un electroencefalograma humano, sentando las bases para el estudio de la actividad eléctrica cerebral. Posteriormente, en 1973, Jacques Vidal formalizó el concepto de BCI al proponer el uso de señales EEG para el control de dispositivos externos. Durante las décadas de 1980 y 1990, investigaciones como las lideradas por Wolpaw en el Instituto Wadsworth demostraron que los sujetos podían aprender a modular voluntariamente sus ritmos sensoriomotores para controlar interfaces, como el movimiento de un cursor en pantalla.

A partir de los años 2000, el campo de las BCI experimentó un crecimiento significativo impulsado por la disponibilidad de bases de datos públicas, la realización de competencias internacionales y el desarrollo de algoritmos de procesamiento y clasificación más robustos. Esto permitió mejorar la precisión y confiabilidad de estos sistemas, facilitando su transición desde entornos experimentales hacia aplicaciones más prácticas.

En la actualidad, las BCI tienen diversas aplicaciones, especialmente en el ámbito clínico y de la rehabilitación. Se emplean en el control de prótesis y dispositivos de asistencia para personas con discapacidades motoras, como pacientes con parálisis o enfermedades neurodegenerativas. Asimismo, permiten el desarrollo de sistemas de comunicación alternativa para individuos con limitaciones severas, y se utilizan en procesos de rehabilitación neurológica mediante técnicas de neuroretroalimentación. Además, han comenzado a explorarse en áreas como el control de dispositivos robóticos, videojuegos, realidad virtual y monitoreo cognitivo.

A futuro, se espera que las BCI evolucionen hacia sistemas más precisos y accesibles, integrando técnicas de inteligencia artificial para mejorar la interpretación de las señales cerebrales, así como el desarrollo de sistemas híbridos que combinen EEG con otras modalidades. Estas mejoras permitirán ampliar su aplicación en medicina, rehabilitación y diversas áreas de la bioingeniería.


### *Consultar sobre actividad cerebral registrada por EEG durante movimientos e imaginación de movimiento de las manos y cómo se hace su análisis en sistemas BCI: Electrodos, ritmos cerebrales de interés*

El **electroencefalograma** (EEG) es una técnica no invasiva que permite registrar la actividad eléctrica del cerebro mediante electrodos ubicados sobre el cuero cabelludo. Estas señales reflejan la actividad sincronizada de poblaciones neuronales y son ampliamente utilizadas en el estudio de procesos motores y cognitivos.

Durante la ejecución de movimientos voluntarios, se produce una activación de la corteza motora que se manifiesta como cambios en la potencia de ciertas bandas de frecuencia del EEG. En particular, los ritmos **mu (8–12 Hz)** y **beta (13–30 Hz)**, asociados a estados de reposo motor, presentan una *disminución de su potencia* cuando el sujeto realiza un movimiento. Este fenómeno se conoce como **desincronización relacionada con eventos** (ERD, Event-Related Desynchronization) y refleja un aumento en la actividad neuronal desorganizada asociada a la ejecución de tareas motoras. De manera complementaria, tras la finalización del movimiento puede observarse un incremento en la potencia de estas bandas, fenómeno conocido como **sincronización relacionada con eventos** (ERS, Event-Related Synchronization), asociado a procesos de recuperación o inhibición cortical.

En el caso de la imaginación de movimiento (motor imagery), aunque no existe ejecución física, se activan regiones similares de la corteza motora, generando patrones de ERD comparables a los del movimiento real, aunque generalmente de menor amplitud. Esta característica es fundamental en sistemas BCI, ya que permite inferir la intención de movimiento a partir de la actividad cerebral sin necesidad de acción muscular.

Desde el punto de vista espacial, estas variaciones se observan principalmente en electrodos ubicados sobre la corteza motora según el sistema internacional 10-20. Los electrodos C3, C4 y Cz son de mayor interés: C3 (hemisferio izquierdo) se asocia al control motor de la mano derecha, C4 (hemisferio derecho) al de la mano izquierda, y Cz a la actividad de la línea media motora.

En cuanto a los ritmos cerebrales, el EEG se clasifica convencionalmente en las bandas delta (0.5–4 Hz), theta (4–8 Hz), alpha (8–12 Hz), beta (13–30 Hz) y gamma (>30 Hz). Es importante notar que el ritmo mu corresponde al ritmo alpha registrado específicamente sobre la corteza sensoriomotora, por lo que comparte su rango de frecuencia (8–12 Hz).

 En el contexto de sistemas BCI orientados al control motor, las bandas mu y beta son las más relevantes debido a su relación directa con la activación e inhibición de la corteza motora. La banda gamma, aunque de interés creciente en investigación, tiene menor uso en BCI clásicas debido a su mayor susceptibilidad al ruido muscular.


###  *Consultar y discutir los flujos de procesamiento propuestos para estos tipos de proyectos y que se puedan realizar en el marco del curso.* 

#### Flujos de procesamiento para señales EEG en sistemas BCI

En la literatura se han propuesto diversos flujos de procesamiento para el análisis de señales EEG en sistemas BCI basados en imaginación de movimiento. 

#### Flujos identificados en la literatura

**Flujo 1: Pipeline clásico basado en análisis espectral (PSD)**

Este es el enfoque más tradicional y ampliamente documentado en BCI motor. Consiste en preprocesamiento de la señal (filtrado pasa-banda, notch y re-referenciación), segmentación por eventos, estimación de la densidad espectral de potencia (PSD) mediante el método de Welch en las bandas de interés (mu y beta), y análisis estadístico comparativo entre condiciones. Su principal ventaja es la transparencia matemática y la interpretación fisiológica directa de los resultados en términos de ERD y ERS.

**Flujo 2: Pipeline basado en filtros espaciales (CSP + clasificador)**

El método de Patrones Espaciales Comunes (CSP, Common Spatial Patterns) es uno de los enfoques más utilizados en competencias BCI. Aplica un filtro espacial que maximiza la varianza de una clase mientras minimiza la de otra, extrayendo características discriminativas entre condiciones. Aunque reporta buena precisión de clasificación, requiere un número suficiente de ensayos por condición y no ofrece una interpretación espectral directa de la actividad cerebral.

**Flujo 3: Pipeline basado en transformadas tiempo-frecuencia (STFT / Wavelet)**

Este enfoque utiliza la Transformada de Fourier de Tiempo Corto (STFT) o la Transformada Wavelet para obtener representaciones tiempo-frecuencia de la señal, permitiendo observar cómo evoluciona la potencia espectral a lo largo del tiempo. Es útil cuando se requiere analizar la dinámica temporal de la ERD/ERS. Sin embargo, implica mayor complejidad computacional e interpretativa respecto al análisis espectral clásico.

**Flujo 4: Pipeline basado en aprendizaje profundo (Deep Learning)**

Los enfoques más recientes utilizan redes neuronales convolucionales (CNN) o modelos híbridos que reciben directamente la señal cruda o representaciones tiempo-frecuencia como entrada, eliminando la necesidad de extracción manual de características. Aunque reportan alta precisión, requieren grandes volúmenes de datos, recursos computacionales elevados y ofrecen baja interpretabilidad de los resultados. 

#### Flujo seleccionado para este proyecto

Para este proyecto se adopta el **Pipeline clásico basado en análisis espectral (PSD)**, por las siguientes razones:

En primer lugar, desde el punto de vista del curso de Bioseñales, cada etapa del flujo está directamente vinculada a los contenidos trabajados: el filtrado digital, el análisis espectral mediante la transformada de Fourier y las herramientas estadísticas sobre señales. 

En segundo lugar, desde el punto de vista fisiológico, la PSD es la herramienta más adecuada dado que los fenómenos de interés — la ERD y la ERS — se manifiestan precisamente como cambios de potencia en bandas específicas del EEG. 

En tercer lugar, a diferencia de los enfoques basados en CSP, transformadas tiempo-frecuencia o aprendizaje profundo, el pipeline espectral no requiere un número elevado de ensayos por sujeto, ni recursos computacionales especializados, ni conocimientos previos de aprendizaje automático, lo que lo hace viable y apropiado para el alcance del proyecto.

En teoría el flujo adoptado comprendería las siguientes etapas: preprocesamiento (filtrado pasa-banda 1–40 Hz, filtro notch, re-referenciación CAR y rechazo de épocas), segmentación por eventos, extracción de la PSD por bandas y canales de interés, y análisis estadístico comparativo entre las cuatro condiciones experimentales.

#### Preprocesamiento

Dado que la señal EEG cruda contiene múltiples fuentes de contaminación, el preprocesamiento busca mejorar la calidad de la señal sin introducir transformaciones complejas. Las etapas aplicadas son:

- **Filtrado pasa-banda (1–40 Hz):** Conserva las frecuencias relevantes del EEG, eliminando la deriva de línea base y el ruido de alta frecuencia no relacionado con los ritmos cerebrales de interés.

- **Filtro Notch (60 Hz):** Suprime la interferencia de la red eléctrica presente en los registros.

- **Re-referenciación por promedio común (CAR):** Mejora la relación señal-ruido espacial restando a cada canal el promedio de todos los canales en cada instante de tiempo.

- **Rechazo de épocas por umbral de amplitud:** Descarta segmentos contaminados por artefactos oculares o musculares de gran magnitud, conservando únicamente los segmentos de señal confiables.

#### Segmentación

Una vez preprocesada la señal continua, los registros se dividen en ventanas temporales llamadas épocas, alineadas a los eventos registrados en el archivo de anotaciones de cada sujeto. Cada época queda etiquetada con la condición experimental correspondiente:

- Movimiento real de mano derecha
- Movimiento real de mano izquierda  
- Imaginación de movimiento de mano derecha
- Imaginación de movimiento de mano izquierda

Esta segmentación es fundamental para comparar condiciones experimentales de manera consistente y aislar la respuesta cerebral asociada a cada tarea.

#### Extracción de características — Densidad Espectral de Potencia (PSD)

Una vez segmentadas las señales, se realiza la extracción de características mediante análisis espectral. La estimación de la **densidad espectral de potencia (PSD)** mediante el método de Welch permite cuantificar la distribución de energía en diferentes bandas de frecuencia del EEG. Este método divide cada época en segmentos con solapamiento, calcula la transformada de Fourier de cada uno y promedia los resultados, obteniendo una estimación estable de la potencia en función de la frecuencia.

La PSD se calcula para los canales de interés sobre la corteza motora (C3, C4, Cz y adyacentes) y se integra en cada banda de frecuencia:

| Banda     | Rango de frecuencia | Relevancia en BCI motor                        |
|-----------|---------------------|------------------------------------------------|
| Delta     | 0.5 – 4 Hz          | Baja relevancia, asociada a sueño profundo     |
| Theta     | 4 – 8 Hz            | Procesos cognitivos y memoria de trabajo       |
| Mu/Alpha  | 8 – 12 Hz           | **Alta relevancia:** ERD/ERS en corteza motora |
| Beta      | 13 – 30 Hz          | **Alta relevancia:** ERD/ERS en corteza motora |
| Gamma     | > 30 Hz             | Procesamiento de alto nivel, susceptible a EMG |

#### Análisis estadístico

Con las características extraídas se procede al análisis comparativo entre las cuatro condiciones experimentales. El flujo incluye:

- 1. **Estadística descriptiva:** Visualización de la distribución de potencia por banda y condición mediante boxplots o violin plots, permitiendo identificar tendencias y diferencias entre hemisferios y condiciones.

- 2. **Prueba de normalidad (Shapiro-Wilk):** Determina si los datos siguen una distribución normal, lo cual define la elección de la prueba estadística.

- 3. **Pruebas de hipótesis:** Según el resultado de la prueba de normalidad, se aplican pruebas paramétricas o no paramétricas identificar si existen diferencias significativas.



#### Discusión del flujo de procesamiento

El flujo de procesamiento descrito corresponde a un enfoque ampliamente utilizado en sistemas BCI basados en EEG, debido a su equilibrio entre complejidad computacional y capacidad de extracción de información relevante.

El uso de técnicas de preprocesamiento como el filtrado pasa-banda y la eliminación de ruido permite mejorar la calidad de la señal sin introducir transformaciones complejas, lo cual resulta adecuado dentro del marco del curso de Bioseñales. Cada etapa está directamente vinculada a los contenidos trabajados: el filtrado digital, el análisis espectral mediante la transformada de Fourier y las pruebas estadísticas sobre señales.

La elección del análisis espectral mediante la PSD responde a su capacidad para caracterizar de manera eficiente la distribución de energía en las bandas de frecuencia de interés, especialmente en los ritmos mu y beta, los cuales están directamente relacionados con la actividad motora cortical. Aunque existen métodos más avanzados, como transformadas tiempo-frecuencia o técnicas basadas en aprendizaje profundo, estos requieren mayor complejidad computacional y una interpretación más elaborada, lo que podría excedre el alcance del curso.

Asimismo, la segmentación de la señal basada en eventos permite realizar comparaciones directas entre condiciones experimentales, facilitando el análisis estadístico posterior. 
Sin embargo, este enfoque depende de una correcta sincronización de los eventos y puede verse afectado por variabilidad entre sujetos.

En este contexto, el flujo propuesto representa una solución adecuada y viable para los objetivos del proyecto, permitiendo analizar y comparar la actividad cerebral asociada a movimiento real e imaginación de movimiento mediante herramientas accesibles y metodologías bien fundamentadas en la literatura de BCI.

```python
# Runs de interés:
# R03, R07, R11 → movimiento real mano derecha e izquierda
# R04, R08, R12 → imaginación de movimiento mano derecha e izquierda
RUNS_INTERES = ['R03', 'R04', 'R07', 'R08', 'R11', 'R12']
```

```python

archivos = [] #lista vacia para las rutas rutas de los archivos

for root, dirs, files in os.walk(RUTA_DATOS): #recorrer las carpetas y subcarperas
    for file in files:  #recorrer cada archivo en la carpeta actual 
        if file.endswith(".edf"): #filtra SOLO archivos EEG
            archivos.append(os.path.join(root, file)) #guardar ruta

print("Archivos encontrados:", len(archivos))
```

```python
archivos_validos = [a for a in archivos 
                    if any(run in a for run in RUNS_INTERES)]
archivos_validos.sort()

print(f"Archivos de runs de interés: {len(archivos_validos)}")
```

```python
import re

RUNS_INTERES = ['R03', 'R04', 'R07', 'R08', 'R11', 'R12']

def run_exacto(ruta, runs):
    nombre = os.path.basename(ruta)  # ej: S001R03.edf
    return any(re.search(rf'{run}\.edf$', nombre) for run in runs)

archivos_validos = [a for a in archivos if run_exacto(a, RUNS_INTERES)]
archivos_validos.sort()

print(f"Archivos de runs de interés: {len(archivos_validos)}")
print(f"Primeros 6: ")
for a in archivos_validos[:6]:
    print(f"  {os.path.basename(a)}")
```

```python
def cargar_sujeto(archivos, sujeto='S001'):
    """
    Carga y concatena todos los runs de interés de un sujeto.
    
    Parámetros:
        archivos (list): Lista de rutas de archivos EDF
        sujeto (str): Nombre del sujeto (ej: 'S001')
    
    Retorna:
        raw (mne.io.Raw): Señal concatenada de todos los runs
    """
    runs_sujeto = [a for a in archivos if sujeto in a]
    runs_sujeto.sort()
    
    print(f"Sujeto {sujeto}: {len(runs_sujeto)} runs encontrados")
    for r in runs_sujeto:
        print(f"  {os.path.basename(r)}")
    
    raws = []
    for ruta in runs_sujeto:
        raw_temp = read_raw_edf(ruta, preload=True, verbose=False)
        mne.datasets.eegbci.standardize(raw_temp)
        raw_temp.set_montage('standard_1005')
        raws.append(raw_temp)
    
    raw = mne.concatenate_raws(raws, verbose=False)
    print(f"Señal concatenada: {raw.times[-1]:.1f} segundos en total")
    return raw
```

```python
def preprocesar_eeg(raw, visualizar=True):
    """
    Recibe una señal EEG cruda y aplica el flujo de preprocesamiento.

    Parámetros:
        raw (mne.io.Raw): Señal EEG cargada
        visualizar (bool): Si True, grafica la señal en cada fase

    Retorna:
        epochs (mne.Epochs): Épocas preprocesadas y etiquetadas
    """

    raw_original = raw.copy()

    # ----------------------------------------------------------
    # FASE 1: Señal cruda
    # ----------------------------------------------------------
    print("=" * 50)
    print("FASE 1: Señal cruda")
    print("=" * 50)
    if visualizar:
        raw_original.plot(
            n_channels=3,
            picks=CANALES_INTERES,
            duration=10,
            title='Fase 1 — Señal cruda',
            scalings='auto'
        )
        plt.show()

    # ----------------------------------------------------------
    # FASE 2: Filtrado pasa-banda + Notch
    # ----------------------------------------------------------
    print("=" * 50)
    print("FASE 2: Filtrado pasa-banda (1–40 Hz) + Notch (60 Hz)")
    print("=" * 50)
    raw.filter(1, 40, verbose=False)
    raw.notch_filter(60, verbose=False)

    if visualizar:
        raw.plot(
            n_channels=3,
            picks=CANALES_INTERES,
            duration=10,
            title='Fase 2 — Después del filtrado',
            scalings='auto'
        )

        # # Comparación espectral antes y después
        # fig, axes = plt.subplots(1, 2, figsize=(12, 4))
        # raw_original.plot_psd(fmax=60, picks=CANALES_INTERES, 
        #                       ax=axes[0], show=False)
        # axes[0].set_title('PSD — Señal cruda')
        # raw.plot_psd(fmax=60, picks=CANALES_INTERES, 
        #              ax=axes[1], show=False)
        # axes[1].set_title('PSD — Después del filtrado')
        # plt.tight_layout()
        # plt.show()

    # ----------------------------------------------------------
    # FASE 3: Re-referenciación por promedio común (CAR)
    # ----------------------------------------------------------
    print("=" * 50)
    print("FASE 3: Re-referenciación por promedio común (CAR)")
    print("=" * 50)
    raw.set_eeg_reference('average', verbose=False)
    raw.apply_proj()

    if visualizar:
        raw.plot(
            n_channels=3,
            picks=CANALES_INTERES,
            duration=10,
            title='Fase 3 — Re-referenciación CAR',
            scalings='auto'
        )
        plt.show()

    # ----------------------------------------------------------
    # FASE 4: Segmentación basada en eventos
    # ----------------------------------------------------------
    print("=" * 50)
    print("FASE 4: Segmentación basada en eventos")
    print("=" * 50)
    eventos, id_eventos = mne.events_from_annotations(raw, verbose=False)
    print(f"  Eventos encontrados: {len(eventos)}")
    print(f"  Condiciones:        {id_eventos}")

    # ----------------------------------------------------------
    # FASE 5: Creación de épocas + rechazo por umbral
    # ----------------------------------------------------------
    print("=" * 50)
    print("FASE 5: Épocas")
    print("=" * 50)
    epochs = mne.Epochs(
        raw,
        eventos,
        event_id=id_eventos,
        tmin=-1,
        tmax=4,
        picks=CANALES_INTERES,
        baseline=(-1, 0),
        reject=dict(eeg=100e-6),
        preload=True,
        verbose=False
    )

    print(f"  Épocas totales:     {len(epochs)}")
    print(f"  Épocas por condición:")
    for cond in epochs.event_id:
        print(f"    {cond}: {len(epochs[cond])}")

    if visualizar:
        epochs.plot(n_epochs=5, title='Fase 5 — Épocas segmentadas')
        plt.show()

    return epochs

#
#  USO EN UN SUJETO
#

# Cargar runs del sujeto S001
raw_sujeto = cargar_sujeto(archivos_validos, sujeto='S001')

# Aplicar preprocesamiento
epochs = preprocesar_eeg(raw_sujeto, visualizar=True)
```
