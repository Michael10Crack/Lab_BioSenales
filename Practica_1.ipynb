## Vectores

```python
import numpy as np

a = np.array([3.1,1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
Mult_escalar = np.dot(a, b) # me arroja un escalar 
print(Mult_escalar)
Multiplicacion = a*b 
# Como los vectores tienen la misma cantidad de componentes, no hace falta modificarlos para poder hacer la multiplicación
# me arroja otro vector de la misma dimension
print(Multiplicacion)

```

## Matrices

```python
A = np.array([  
            [2      , -1    , -3    ]   ,
            [4      , 1.5   , -2.5  ]   ,
            [7.3    , -0.9  , 0.2   ]   ,
])

Trasnpuesta = np.transpose(A)

print(Trasnpuesta)
print("---------"*5)

unos = np.ones((2,3)) #crea una matriz de 1's de la dimension especificada
print("Unos" , '\n', unos)
print("---------"*5)

round = A # poodemos usar nuestra matriz original
# el comando redondeará los numeros de la matriz al entero más cercano
print("Redondeo" , "\n" , np.round(round))
print("---------"*5)

redondeo_up = A #este comando redondea al numero entero mayor más cercano
print("Redondeo para arriba" , "\n" , np.ceil(redondeo_up))
print("---------"*5)

redondeo_down = A #este comando redondea al numero entero menor más cercano
print("Redondeo para abajo" , "\n" , np.floor(redondeo_down))
print("---------"*5)


#Casilla primera fila tercera columna
print(A)
print("Valor primera fila, tercera columna:" , A[0,2])
print("---------"*5)

#Toda una fila
print("Segunda fila:", A[1 ,:])
print("---------"*5)

#dimension de una matriz
print("Dimension de la matriz:" , A.shape)


```

## Señales

```python
#creacion de la funcion
#primero crear el intervalo
n = np.arange(0 , 101)
# ahora la funcion como tal
y = np.sin(np.pi * 0.12* n )

# la otra funcion 
z = np.cos(2*np.pi * 0.03*n)

#aunque para este ejemplo ambas señales tienen el mismo intervalo n es una buena practica verficar que ambas señales sean del mismo intervalo
print( y.shape  , z.shape )

#sumemos
Suma_senales = y + z

#multipliquemos
Multiplicacion_senales = y * z

#graficar 
import matplotlib.pyplot as plt 
plt.figure()
plt.plot(n , y , label= "y[n] = seno", color = "blue")
plt.plot(n , z , label= "y2[n] = cos", color = "red")
plt.xlabel("Periodo ")
plt.ylabel("Amplitud")
plt.title("Señales Originales")
plt.legend()
plt.grid(False) # no me gusta el grid
plt.show()

plt.figure()
plt.plot(n, Suma_senales , label="s[n] = Suma" , color="purple")
plt.plot(n , Multiplicacion_senales , label = "t[n] = Mult." , color = "black")
plt.xlabel("Periodo")
plt.ylabel("Amplitud")
plt.title("Señales Operadas")
plt.legend()
plt.grid(False)
plt.show()
```

## Pandas

```python
import pandas as pd
#la estructura de los diccionarios constan de "clave" : valor
Estudiantes = {
    "Miguel"    : 1.0,
    "Andres"    : 2.9,
    "Ciara"     : 5.0,
    "Maria"     : 3.2,
    "Jose"      : 4.5,
    "Catalina"  : 2.2
}

#Ahora toca convertir el diccionario a una serie
serie = pd.Series(Estudiantes)
#print(serie)
#ahora la estadística
#crear los resultados como series
resultados = pd.Series({
"Minimo" : serie.min(),
"Maximo" : serie.max(),
"Mediana": serie.mean(),
"Desviación estándar": serie.std()

})
print(resultados)
```

## Cargar datos CSV


```python
df = pd.read_csv("datos.csv")
first = df.head()
last = df.tail() 
print("Primeras filas del DataFrame:" , "\n" , first)
print("Ultimas filas del DataFrame:" , "\n" , last)

df = df.drop(columns=["Unnamed: 0"])
print("Primeras filas del DataFrame:" , "\n" , first)
print("Ultimas filas del DataFrame:" , "\n" , last)
# aquií hay un problema y es que este archivo csv no está separado por columnas, sino que está separado por ; por lo que lo ideal sería 
# especificar el separador al momento de leer el archivo csv
```

```python
df_correcto = pd.read_csv("datos.csv" , sep=";")
print("Primeras filas del DataFrame:" , "\n" , df_correcto.head())
print("Ultimas filas del DataFrame:" , "\n" , df_correcto.tail())
# ahora sí se ve como debería ser
df = df_correcto.drop(columns=["Unnamed: 0"])
print("-----------------"*5)
print(df.head())
print(df.tail())
# otro problema ahora es que según parece, están trocadas los datos de las columnas weight y height, 
# #lo que se puede solucionar fácilmente intercambiando los nombres de las columnas
df["Weight"], df["Height"] = df["Height"], df["Weight"]
print("------------------"*5)
print(df.head())
#ahora sí podemos proceder a calcular el IMC
print("------------------"*5)
df["BMI"] = df["Weight"] / (df["Height"]/100)**2
print(df.head())

#extra
import numpy as np
df["Categoría"] = np.select(
    [
        df["BMI"] < 18.5,
        df["BMI"] < 25,
        df["BMI"] < 30
    ],
    [
        "Bajo peso",
        "Normal",
        "Sobrepeso"
    ],
    default="Obesidad"
)
print(df)
```
