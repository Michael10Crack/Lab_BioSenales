```python
import numpy as np
import matplotlib.pyplot as plt
def impseq(n0, n1, n2):
    # Genera x(n) = delta(n-n0); n1 <= n <= n2
    # ----------------------------------------------
    n = np.arange(n1,n2+1) # Se crea el vector de muestras
    x = (n-n0) == 0
    return [x,n]

[x,n] = impseq(-1, -5, 5)

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Impulso unitario')
plt.grid(True)
plt.show()
```

```python
def escalon(n0, n1, n2):
    # Genera x(n) = delta(n-n0); n1 <= n <= n2
    # ----------------------------------------------
    n = np.arange(n1,n2+1) # Se crea el vector de muestras
    x = (n >= n0) == 1
    
    
    return [x,n]

[x,n] = escalon(4, -5, 5)

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('escalon unitario')
plt.grid(True)
plt.show()
```

```python
def rampa(n0, n1, n2):
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n - n0) * (n >= n0)      # condicion para que sea 1 o True
 return [x,n]

#Ejemplo
x, n = rampa(-3, -4, 4)
plt.stem(n, x)
plt.title("Rampa r(n)")
plt.xlabel("n")
plt.ylabel("r(n)")
plt.grid(True)
plt.show()
```

```python
# Exponencial compleja
def exponencial_compleja(n0, n1, n2, w0, sigma):
    n = np.arange(n1, n2+1)
    x = np.exp(sigma + 1j * w0 * (n - n0))
    return n, x

n, x = exponencial_compleja(0, -5, 5, 0, 0.01)

plt.stem(n, np.real(x))
plt.xlabel('n')
plt.ylabel('Re{x(n)}')
plt.title('Parte Real de la Exponencial Compleja')
plt.grid(True)
plt.show()
```

```python


# Eje temporal
n = np.arange(0,18)

# Inicializar señal
x = np.zeros(len(n))

# Definición por tramos
for i in range(len(n)):
    if 0 <= n[i] <= 5:
        x[i] = n[i]
    elif 6 <= n[i] <= 11:
        x[i] = n[i] - 5
    elif 12 <= n[i] <= 17:
        x[i] = n[i] - 10

# Derivada discreta
y = np.zeros(len(n))
y[1:] = x[1:] - x[:-1]

# Subplot
plt.figure()

plt.subplot(2,1,1)
plt.stem(n,x)
plt.title('Secuencia original x[n]')
plt.grid()

plt.subplot(2,1,2)
plt.stem(n,y)
plt.title('Derivada discreta y[n]')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
def impseq(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n == n0)
    return x.astype(float), n

def escalon(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n >= n0)
    return x.astype(float), n

def rampa(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n - n0) * (n >= n0)
    return x.astype(float), n




n = np.arange(-5,19)   

#rampa creciente hasta n=0
r1, _ = rampa(-5, -5, 0)   
r1 = r1

#  rampa decreciente
r2 = np.array([4,3,2,1,0])

# rampa creciente pequeña
r3 = np.array([1,2,3,4,5])

# constante 5
c1 = 5*np.ones(4)

# constante 10
c2 = 10*np.ones(4)

x = np.concatenate((r1, r2, r3, c1, c2))

plt.figure()
plt.stem(n, x)
plt.title("Secuencia original x(n)")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(True)
plt.show()


y1 = np.zeros(len(n))

for i in range(len(n)):
    
    # x(n)
    xn = x[i]
    
   
    if -5 <= n[i]-4 <= 18:
        xn_shift = x[(n[i]-4) - (-5)]
    else:
        xn_shift = 0
    
    y1[i] = 2*xn_shift + xn

plt.figure()
plt.stem(n, y1)
plt.title("X_5(n) = 2x(n-4) + x(n)")
plt.xlabel("n")
plt.ylabel("X_5(n)")
plt.grid(True)
plt.show()


n2 = np.arange(-20,21)
y2 = np.zeros(len(n2))

for i in range(len(n2)):
    
    # x(n)
    if -5 <= n2[i] <= 18:
        xn = x[n2[i] - (-5)]
    else:
        xn = 0
    
    
    if -5 <= n2[i]+2 <= 18:
        xn_shift2 = x[(n2[i]+2) - (-5)]
    else:
        xn_shift2 = 0
    
    y2[i] = 0.001*np.exp(0.5*n2[i])*xn+10*np.sin(0.05*np.pi*n2[i])*xn_shift2

plt.figure()
plt.stem(n2, y2)
plt.title("X_6(n)  -20 ≤ n ≤ 20")
plt.xlabel("n")
plt.ylabel("X_6(n)")
plt.grid(True)
plt.show()
```
