```python
import numpy as np
import matplotlib.pyplot as plt
def impseq(n0, n1, n2):
    # Genera x(n) = delta(n-n0); n1 <= n <= n2
    # ----------------------------------------------
    n = np.arange(n1,n2+1) # Se crea el vector de muestras
    x = (n-n0) == 0
    return [x,n]

[x,n] = impseq(-1, -5, 5)

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Impulso unitario')
plt.grid(True)
plt.show()
```

```python
def escalon(n0, n1, n2):
    # Genera x(n) = delta(n-n0); n1 <= n <= n2
    # ----------------------------------------------
    n = np.arange(n1,n2+1) # Se crea el vector de muestras
    x = (n >= n0) == 1
    
    
    return [x,n]

[x,n] = escalon(4, -5, 5)

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('escalon unitario')
plt.grid(True)
plt.show()
```

```python
def rampa(n0, n1, n2):
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n - n0) * (n >= n0)      # condicion para que sea 1 o True
 return [x,n]

#Ejemplo
x, n = rampa(-3, -4, 4)
plt.stem(n, x)
plt.title("Rampa r(n)")
plt.xlabel("n")
plt.ylabel("r(n)")
plt.grid(True)
plt.show()
```

```python
# Exponencial compleja
def exponencial_compleja(n0, n1, n2, w0, sigma):
    n = np.arange(n1, n2+1)
    x = np.exp(sigma + 1j * w0 * (n - n0))
    return n, x

n, x = exponencial_compleja(0, -5, 5, 0, 0.01)

plt.stem(n, np.real(x))
plt.xlabel('n')
plt.ylabel('Re{x(n)}')
plt.title('Parte Real de la Exponencial Compleja')
plt.grid(True)
plt.show()
```

```python
#operacion de impulsos
x1n , n = impseq(-1, -5, 5)
x2n , _= impseq(-3, -5, 5)
x3n , _ = impseq(-2, -5, 5)
x4n , _ = impseq(0, -5, 5)
x5n , _ = impseq(0, -5, 5)
x = 3*x1n + 5*x2n + 3*x3n + 3*x4n + x5n
 

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x1(n)')
plt.title("impulsos")
plt.show()
```

```python
#operacion de escalones
un1 , n = escalon(-4, -6, 6)
un2 , _ = escalon(-2, -6, 6)

w = -2*un1 + 3*un2
plt.stem(n,w)
plt.xlabel('n')
plt.ylabel('w(n)')  
plt.title("escalones")
plt.show()  
```

```python
n = np.arange(0 , 101)
x = np.exp(0.01*n) * np.sin(2*np.pi*n)
plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('exponencial modulada por sinusoidal')
plt.grid(True)
plt.show()
```

```python
x4n , n = rampa(-3 , -10, 10)
x5n , _ = rampa(2 , -10, 10)
x6n , _ = escalon(3 , -10, 10)
x = 2*x4n - 1*x5n - 5*x6n
plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title("rampas y escalones")
plt.grid(True)
plt.show()
```

```python


# Eje temporal
n = np.arange(0,18)

# Inicializar seal
x = np.zeros(len(n))

# Definicin por tramos
for i in range(len(n)):
    if 0 <= n[i] <= 5:
        x[i] = n[i]
    elif 6 <= n[i] <= 11:
        x[i] = n[i] - 5
    elif 12 <= n[i] <= 17:
        x[i] = n[i] - 10

# Derivada discreta
y = np.zeros(len(n))
y[1:] = x[1:] - x[:-1]

# Subplot
plt.figure()

plt.subplot(2,1,1)
plt.stem(n,x)
plt.title('Secuencia original x[n]')
plt.grid()

plt.subplot(2,1,2)
plt.stem(n,y)
plt.title('Derivada discreta y[n]')
plt.grid()

plt.tight_layout()
plt.show()
```
